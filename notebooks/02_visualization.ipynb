```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent
sys.path.insert(0, str(PROJECT_ROOT))
```

```python
import pandas as pd
from src.db import get_engine
```

```python
engine = get_engine()
```

```python
query = """
SELECT * 
FROM
credit_card_transactions;
"""
```

```python
df = pd.read_sql(query, engine)
```

```python
df.head()
```

```python
df.tail()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(6, 4))

sns.countplot(
    data=df,
    x="class"
)

plt.title("Class Distribution")
plt.xlabel("Class")
plt.ylabel("Count")

plt.savefig(
    "/home/harsh/Documents/credit-card-fraud-detection/figures/seaborn/01_class_distribution.png",
    dpi=300,
    bbox_inches="tight"
)

plt.close()
```

```python
class_percent = (
    df["class"]
    .value_counts(normalize=True)
    .mul(100)
    .round(3)
)

print(class_percent)
```

```python
plt.figure(figsize=(6, 6))

class_percent.plot(
    kind="pie",
    autopct="%1.3f%%",
    ylabel="",
    startangle=90
)

plt.title("Percentage of Legitimate vs Fraud Transactions")

plt.savefig(
    "/home/harsh/Documents/credit-card-fraud-detection/figures/seaborn/02_class_percentage.png",
    dpi=300,
    bbox_inches="tight"
)

plt.close()
```

```python
plt.figure(figsize=(10, 5))

sns.histplot(
    data=df,
    x="amount",
    bins=50,
    kde=True
)

plt.title("Distribution of Transaction Amount")
plt.xlabel("Transaction Amount")
plt.ylabel("Frequency")

plt.savefig(
    "/home/harsh/Documents/credit-card-fraud-detection/figures/seaborn/03_transaction_amount_distribution.png",
    dpi=300,
    bbox_inches="tight"
)

plt.close()
```

```python
plt.figure(figsize=(10, 2.5))

sns.boxplot(
    data=df,
    x="amount"
)

plt.title("Transaction Amount Boxplot")

plt.savefig(
    "/home/harsh/Documents/credit-card-fraud-detection/figures/seaborn/04_transaction_amount_boxplot.png",
    dpi=300,
    bbox_inches="tight"
)

plt.close()
```

```python
plt.figure(figsize=(10, 6))

sns.boxplot(
    data=df,
    x="class",
    y="amount"
)

plt.title("Transaction Amount by Class")
plt.xlabel("Class")
plt.ylabel("Transaction Amount")

plt.savefig(
    "/home/harsh/Documents/credit-card-fraud-detection/figures/seaborn/05_amount_by_class_boxplot.png",
    dpi=300,
    bbox_inches="tight"
)

plt.close()
```

```python
plt.figure(figsize=(10, 6))

sns.violinplot(
    data=df,
    x="class",
    y="amount",
    inner="quartile"
)

plt.title("Transaction Amount Distribution by Class")
plt.xlabel("Class")
plt.ylabel("Transaction Amount")

plt.savefig(
    "/home/harsh/Documents/credit-card-fraud-detection/figures/seaborn/06_amount_by_class_violin.png",
    dpi=300,
    bbox_inches="tight"
)

plt.close()
```

```python
plt.figure(figsize=(16, 12))

corr = df.corr(numeric_only=True)

sns.heatmap(
    corr,
    cmap="coolwarm",
    center=0
)

plt.title("Correlation Heatmap")

plt.savefig(
    "/home/harsh/Documents/credit-card-fraud-detection/figures/seaborn/07_correlation_heatmap.png",
    dpi=300,
    bbox_inches="tight"
)

plt.close()
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(
    data=df,
    x="amount",
    hue="class",
    bins=50,
    kde=True,
    stat="density",
    common_norm=False
)

plt.title("Transaction Amount Distribution by Class")

plt.savefig(
    "/home/harsh/Documents/credit-card-fraud-detection/figures/seaborn/08_amount_distribution_by_class.png",
    dpi=300,
    bbox_inches="tight"
)

plt.close()
```

```python
import plotly.express as px
```

```python
class_counts = (
    df["class"]
    .value_counts()
    .rename_axis("Class")
    .reset_index(name="Count")
)

fig = px.bar(
    class_counts,
    x="Class",
    y="Count",
    color="Class",
    text="Count",
    title="Class Distribution"
)

fig.update_layout(
    template="plotly_white",
    xaxis_title="Class",
    yaxis_title="Number of Transactions"
)

fig.write_html("/home/harsh/Documents/credit-card-fraud-detection/figures/plotly/01_class_distribution.html")
```

```python
class_percent = (
    df["class"]
    .value_counts(normalize=True)
    .mul(100)
    .round(3)
    .rename_axis("Class")
    .reset_index(name="Percentage")
)

fig = px.pie(
    class_percent,
    names="Class",
    values="Percentage",
    title="Fraud vs Legitimate Transactions"
)

fig.update_traces(textposition="inside", textinfo="percent+label")

fig.write_html("/home/harsh/Documents/credit-card-fraud-detection/figures/plotly/02_class_percentage.html")
```

```python
fig = px.histogram(
    df,
    x="amount",
    nbins=50,
    title="Transaction Amount Distribution"
)

fig.update_layout(
    template="plotly_white",
    xaxis_title="Transaction Amount",
    yaxis_title="Frequency"
)

fig.write_html("/home/harsh/Documents/credit-card-fraud-detection/figures/plotly/03_amount_distribution.html")
```

```python
fig = px.box(
    df,
    x="class",
    y="amount",
    color="class",
    title="Transaction Amount by Class"
)

fig.update_layout(
    template="plotly_white",
    xaxis_title="Class",
    yaxis_title="Transaction Amount"
)

fig.write_html("/home/harsh/Documents/credit-card-fraud-detection/figures/plotly/04_amount_by_class_boxplot.html")
```

```python
import plotly.graph_objects as go

corr = df.corr(numeric_only=True)

fig = go.Figure(
    data=go.Heatmap(
        z=corr.values,
        x=corr.columns,
        y=corr.columns,
        colorscale="RdBu",
        zmid=0
    )
)

fig.update_layout(
    title="Correlation Heatmap",
    template="plotly_white",
    width=900,
    height=900
)

fig.write_html("/home/harsh/Documents/credit-card-fraud-detection/figures/plotly/05_correlation_heatmap.html")
```
